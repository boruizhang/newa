```python
import itertools
verblist = []
with open('for_ml/train.txt') as data:
    for line1,line2 in itertools.zip_longest(*[data]*2):
        verbs = line2.strip().split(",")
        verblist.append(verbs)
with open('for_ml/dev.txt') as data:
    for line1,line2 in itertools.zip_longest(*[data]*2):
        verbs = line2.strip().split(",")
        verblist.append(verbs)
with open('for_ml/test.txt') as data:
    for line1,line2 in itertools.zip_longest(*[data]*2):
        verbs = line2.strip().split(",")
        verblist.append(verbs)
```

```python
len(verblist)
```

```python
from collections import defaultdict
matrixverbs = defaultdict(int)
embeddedverbs = defaultdict(int)
for verbs in verblist:
    print(verbs)
    try:
        m, e = verbs
    except ValueError:
        continue
    matrixverbs[m] += 1
    embeddedverbs[e] += 1
```

```python
lines = []
for line in open("sp1/0_100_dhaka_true_withEn.txt"):
    lines.append(line.strip())
for line in open("sp1/500_600_dhaka_true_withEn.txt"):
    lines.append(line.strip())
```

```python
sentences = lines[0::3]
nverbs = lines[1::3]
everbs = lines[2::3]
```

```python
import re
newari_mverbs = []
newari_everbs = []
english_mverbs = []
english_everbs = []
translation_dict = defaultdict(list)
for nv, ev in zip(nverbs, everbs):
    nmatch = re.match(r'\[(.*),(.*)\]', nv)
    ematch = re.match(r'\[(.*),(.*)\]', ev)
    try:
        nmv = nmatch.group(1).strip()
        nev = nmatch.group(2).strip()
        emv = ematch.group(1).strip()
        eev = ematch.group(2).strip()
        
    except AttributeError:
        continue
    newari_mverbs.append(nmv)
    newari_everbs.append(nev)
    english_mverbs.append(emv)
    english_everbs.append(eev)
    translation_dict[nmv].append(emv)
    translation_dict[nev].append(eev)
```

```python
translation_dict
```

```python
for x in sorted(matrixverbs, key=lambda x: matrixverbs[x], reverse=True):
    print(x, matrixverbs[x], translation_dict[x])
```

```python
for x in sorted(embeddedverbs, key=lambda x: embeddedverbs[x], reverse=True):
    print(x, embeddedverbs[x], translation_dict[x])
```
