```python
from collections import defaultdict
import itertools
# import WhitespaceTokenizer() method from nltk
from nltk.tokenize import WhitespaceTokenizer
d = defaultdict(list)
d['sentences'] = []
d['tags'] = []


# Create a reference variable for Class WhitespaceTokenizer
tk = WhitespaceTokenizer()
with open('dev.txt') as f: # runt on dev and train txt file separately
    for line1,line2 in itertools.zip_longest(*[f]*2):
        sents = []
        tags = []
        tokens = iter(tk.tokenize(line1))
        current_tag = ""
        for token in tokens:
            if token in line2:    
                # Good: even if the verb is complex (aka spaced out),suffix can still apply to the entire phrase
                # Bad: false positive -V may occur, if one word that has multiple meanings occurring more than one time in the same sentence
                suffix = "-V"
            else:
                suffix = ""
            try: 
                if token == "[":
                    current_tag = "IEC"
                    tags.append("BEC"+suffix)
                    sents.append(next(tokens))
                elif token == "]":
                    current_tag = ""
            #        tags.append(current_tag+suffix)
            #        sents.append(next(tokens))
                else:
                    tags.append(current_tag+suffix)
                    sents.append(token)
            except StopIteration:
                pass
        d['sentences'].append(sents)
        d['tags'].append(tags)
        
        
```

```python
from collections import defaultdict
import itertools
# import WhitespaceTokenizer() method from nltk
from nltk.tokenize import WhitespaceTokenizer
t = defaultdict(list)
t['sentences'] = []
t['tags'] = []


# Create a reference variable for Class WhitespaceTokenizer
tk = WhitespaceTokenizer()
with open('train.txt') as f: # runt on dev and train txt file separately
    for line1,line2 in itertools.zip_longest(*[f]*2):
        sents = []
        tags = []
        tokens = iter(tk.tokenize(line1))
        current_tag = ""
        for token in tokens:
            if token in line2:    
                # Good: even if the verb is complex (aka spaced out),suffix can still apply to the entire phrase
                # Bad: false positive -V may occur, if one word that has multiple meanings occurring more than one time in the same sentence
                suffix = "-V"
            else:
                suffix = ""
            try: 
                if token == "[":
                    current_tag = "IEC"
                    tags.append("BEC"+suffix)
                    sents.append(next(tokens))
                elif token == "]":
                    current_tag = ""
            #        tags.append(current_tag+suffix)
            #        sents.append(next(tokens))
                else:
                    tags.append(current_tag+suffix)
                    sents.append(token)
            except StopIteration:
                pass
        t['sentences'].append(sents)
        t['tags'].append(tags)
        
```

```python
import NERDA
training = d
validation = t
```

```python
print(training.get('sentences')[0])
```

```python
tag_scheme = [
'BEC',
'BEC-V',
'IEC',
'IEC-V',
'-V',
''
]
```

```python
from NERDA.models import NERDA
model = NERDA(
dataset_training = training,
dataset_validation = validation,
tag_scheme = tag_scheme, 
tag_outside = 'O',
transformer = 'bert-base-multilingual-cased',
hyperparameters = {'epochs' : 1,
                  'warmup_steps' : 10,
                  'train_batch_size': 5,
                  'learning_rate': 0.0001})
```

```python
model.train()
```

```python
model.evaluate_performance(validation)
```
